```python
import matplotlib.pyplot as plt
import re

file_path='log.txt'
with open(file_path, 'r') as file:
    lines = file.readlines()

# 初始化列表以存储数据
epochs = []
dis_losses = []
gen_losses = []
identity_losses = []

for i in range(0, 6500, 50):
    line = lines[i]
    # 使用正则表达式提取数字
    values = re.findall(r"[-+]?\d*\.\d+|\d+", line)

    # 提取相关数据
    epoch = int(values[1])
    dis_loss = float(values[2])
    gen_loss = float(values[3])
    identity_loss = float(values[4])
    epochs.append(i * 50)  # 使用索引乘以50作为横坐标

    # 将数据添加到相应的列表中
    epochs.append(epoch)
    dis_losses.append(dis_loss)
    gen_losses.append(gen_loss)
    identity_losses.append(identity_loss)
    x_values = [i * 50 for i in range(len(dis_losses))]


plt.figure(figsize=(10, 6))
plt.plot(x_values, dis_losses, label='DIS loss', color='green')
plt.plot(x_values, gen_losses, label='GEN loss',  color='dodgerblue')
plt.plot(x_values, identity_losses, label='Identity loss', color='orange')

# 添加标题和标签
plt.title('Training Losses Over Epochs (Every 50th Line)')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()  # 显示图例

# 显示图形
plt.show()


```
